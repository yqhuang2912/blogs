```python
import numpy as np
import matplotlib.pyplot as plt

# Example logits (from your blog example)
z_m = 5.2  # max logit (e.g., "物理学家")
z_k = 5.0  # runner-up logit (e.g., "数学家")
delta = z_k - z_m  # negative

taus = np.linspace(0.05, 3.0, 500)
ratio = np.exp(delta / taus)

plt.figure(figsize=(7.5, 4.8))
plt.plot(taus, ratio)
plt.xlabel(r"temperature $\tau$")
plt.ylabel(r"$\frac{p_{\tau}(k)}{p_{\tau}(m)} = \exp\left(\frac{z_k - z_m}{\tau}\right)$")
plt.title(r"Effect of Temperature on Relative Odds (fixed $\Delta=z_k-z_m<0$)")
plt.yscale("log")  # shows the change more clearly
plt.grid(True, which="both", linestyle="--", linewidth=0.6, alpha=0.7)
plt.tight_layout()
plt.show()

# Also plot for multiple gaps to show generality
deltas = np.array([-0.1, -0.5, -1.0, -2.0])
plt.figure(figsize=(7.5, 4.8))
for d in deltas:
    plt.plot(taus, np.exp(d / taus), label=rf"$\Delta=z_k-z_m={d}$")
plt.xlabel(r"temperature $\tau$")
plt.ylabel(r"$\exp(\Delta/\tau)$")
plt.title("Relative Odds vs Temperature for Different Logit Gaps")
plt.yscale("log")
plt.grid(True, which="both", linestyle="--", linewidth=0.6, alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
